## Load Data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import operator
import numpy as np
```

```python
with open('Data.txt') as f:
    beats = f.read().split()
beats = [float(b) for b in beats]
```

## Plot ECG

```python
plt.figure(figsize=(15,8))
plt.plot(beats)
plt.ylabel("ECG (mV)")
plt.xlabel("Time (sec)")
```

## Finding R-Peaks

```python
maxima = []

# Gather all local maxima
for x, y in enumerate(beats):
    if x > 0 and x < len(beats)-1:
        if y > beats[x-1] and y > beats[x+1]:
            maxima.append((x, y))
            
# Sort the ECG values for all maxima in descending order
maxima.sort(key=operator.itemgetter(1), reverse=True)

# Compute the difference between each maxima and the absolute maxima (highest R-peak)
# When the difference gets higher than 0.5, we know it's not an R-peak
for i in range(1, len(maxima)):
    if abs(maxima[i][1] - maxima[0][1]) > 0.5:
        maxima = maxima[0:i]
        break
        
# Sort it back by index (time value)
maxima.sort(key=operator.itemgetter(0))

# Now we have our list of R-peaks
print maxima
```

## Calculating R-R Intervals

```python
# Printing absolute differences (in seconds) between each R-peak
r_r_intervals = [abs(maxima[i][0] - maxima[i-1][0]) for i in range(1, len(maxima))]
print r_r_intervals
```

```python
heart_rate = ([60*100/x for x in r_r_intervals])
average_bpm = np.mean(heart_rate)
sd = np.std(heart_rate)
print 'Average heart rate:', average_bpm
print 'Standard Deviation:', sd
```
